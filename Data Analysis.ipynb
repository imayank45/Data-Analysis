```python
import pandas as pd
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
df
```

## How many free apps are there in ART_AND_DESIGN ?

```python
df = df.values
```

```python
c = 0
for i in df:
    if i[1] == 'ART_AND_DESIGN' and i[6] == 'Free':
        c += 1
c
```

## How many paid apps are there in ART_AND_DESIGN ?

```python
c = 0
for i in df:
    if i[1] == 'ART_AND_DESIGN' and i[6] == 'Paid':
        c += 1
c
```

## How many apps are there in ART_AND_DESIGN with rating more than 4.5 ?

```python
c = 0
for i in df:
    if i[1] == 'ART_AND_DESIGN' and i[2] > 4.5:
        c += 1
c
```

## How many apps are there in FAMILY category with rating more than 4.5 and should be free?

```python
ct = 0
for i in df:
    if i[1] == 'FAMILY' and i[2] > 4.5 and i[6] == 'Free':
        ct+= 1
ct
```

## List all apps with rating more than 4.5  in FAMILY category ?

```python
c = 0
for i in df:
    if i[1] == 'FAMILY' and i[2] > 4.5:
        print(i[0])
```

## Name the apps with rating more than 4.5 in ART_AND_DESIGN in ascending order ?

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
df_pr = df_pr[df_pr['Rating'] > 4.5]
df_pr.sort_values(by = 'Rating', ascending = True)
```
